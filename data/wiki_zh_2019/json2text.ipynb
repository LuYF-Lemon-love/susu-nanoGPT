## 导入第三方库

```python
import os
import json
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace
```

## 读取 JSONL 文件

```python
def read_jsonl(jsonl_file):
    text_list = []
    with open(jsonl_file, "r") as file:
        for line in file:
            json_obj = json.loads(line)
            text_list.append(json_obj["text"].lstrip(json_obj["title"]).strip())
    return text_list
```

```python
# 读取当前目录下的所有 excel 文件
folder_path = 'wiki_zh'
text_list = []
for root, dirs, files in os.walk(folder_path):
    for file in files:
        file_path = os.path.join(root, file)
        print(file_path)
        text_list += read_jsonl(file_path)
print(f'一共{len(text_list)}条数据。')
```

```python
text_list[0]
```

```python
txt_path = "all_5%.txt"
part = round(0.05 * len(text_list))
with open(txt_path, 'w', encoding='utf-8') as fout:
    for text in text_list[:part]:
        fout.write(text)
        fout.write("\n\n")
```

```python
txt_path = "all.txt"
with open(txt_path, 'w', encoding='utf-8') as fout:
    for text in text_list:
        fout.write(text)
        fout.write("\n\n")
```

## 分词

```python
tokenizer = Tokenizer(BPE(unk_token="[UNK]"))
trainer = BpeTrainer(special_tokens=["[UNK]", "[CLS]", "[SEP]", "[PAD]", "[MASK]"])
tokenizer.pre_tokenizer = Whitespace()
files = [txt_path]
tokenizer.train(files, trainer)
```

```python
tokenizer.save("tokenizer.json")
```
